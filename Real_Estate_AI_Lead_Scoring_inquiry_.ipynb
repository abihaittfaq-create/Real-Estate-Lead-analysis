```python
import pandas as pd
import json
```

```python
with open("inquiries (1).json", "r") as f:
    data = json.load(f)

df = pd.DataFrame(data)

df.head()
```

```python
print(df.info())
print(df.head())
```

```python
def get_intent(message):

    message = message.lower()

    if "ready to buy" in message:
        return "Buying Intent"

    elif "need more details" in message:
        return "Information Request"

    elif "contact me" in message:
        return "Contact Request"

    elif "exploring options" in message:
        return "Browsing"

    else:
        return "Other"
```

```python
df["Intent"] = df["message"].apply(get_intent)

df[["message","Intent"]].head(10)
```

```python
def lead_score(row):

    score = 0

    # Timeline Score
    if row["timeline"] == "Within 1 month":
        score += 40

    elif row["timeline"] == "Within 3 months":
        score += 25

    elif row["timeline"] == "6 months":
        score += 15

    else:
        score += 5

    # Intent Score
    if row["Intent"] == "Buying Intent":
        score += 40

    elif row["Intent"] == "Contact Request":
        score += 30

    elif row["Intent"] == "Information Request":
        score += 20

    elif row["Intent"] == "Browsing":
        score += 10

    return score
```

```python
df["Lead Score"] = df.apply(lead_score, axis=1)

df.head()
```

```python
def lead_category(score):

    if score >= 70:
        return "Hot Lead"

    elif score >= 40:
        return "Warm Lead"

    else:
        return "Cold Lead"
```

```python
df["Lead Category"] = df["Lead Score"].apply(lead_category)

df.head(20)
```

```python
df[["inquiryId",
    "message",
    "Intent",
    "timeline",
    "Lead Score",
    "Lead Category"]]
```

```python
df.to_csv("Lead_Scoring_Output.csv", index=False)

print("Done!")
```

```python
# User Input Testing

user_message = input("Enter your inquiry: ")

# Intent Recognition
user_intent = get_intent(user_message)

# User details
user_budget = int(input("Enter your budget: "))
user_timeline = input("Enter your timeline: ")
user_location = input("Enter your location: ")


# Create user data
user_data = {
    "message": user_message,
    "budget": user_budget,
    "timeline": user_timeline,
    "location": user_location,
    "Intent": user_intent
}


# Lead Score
user_score = lead_score(user_data)

# Lead Category
user_category = lead_category(user_score)


# Final Result
print("\n--- Lead Analysis ---")
print("Message:", user_message)
print("Intent:", user_intent)
print("Lead Score:", user_score)
print("Lead Category:", user_category)
```
